```python
import pandas as pd
```

```python
df = pd.read_csv('streamverse_dataset_estudo.csv', sep=';') 
df.head(120)
```

Determinar qual Pais teve a maior duração_minutos total de visualização e apresentear visualmente atraves de uma tabela 

```python
df_filter = df.groupby('pais')['duracao_minutos'].sum().reset_index().sort_values('duracao_minutos',ascending=False).head()
df_filter
```

Suponha que o Streamverse deseja implementar um sistema de bônus para usuarios de alta fidelidade a regra é a seguinte:

Se a duracao_minutos media diaria de um usuario for maior que 180 minutos, o usuario recebera um bonus de 20%
 
Se a duracao_minutos media diaria de um usuario for menor ou igual a 180 minutos, o usuario recebera um bonus de 5%

Determine quantos usuarios se qualificariam para o bonus de 20%

```python
df_filter= df.groupby('id_usuario') ['duracao_minutos'].mean().reset_index()
df_maior= df_filter[df_filter['duracao_minutos']>180]
df_resultado= df_maior['id_usuario'].count()
df_resultado
```

Determinar a duracao_minutos total dee visaulização por tipo_plano e apresentar esses restultados viasualmente atravez de uma tabela
