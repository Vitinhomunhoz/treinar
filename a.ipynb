Maior venda do mês "Mar"




```python
df_filter = df[df['mes'] == 'Mar']
df_filter['faturamento'].max()
```

Vendas Totais do mês de "Jan"


```python
df_filter = df[df['mes'] == 'Jan']
df_filter ['faturamento'].sum()
```

Faturamento > 2000


```python
df_filter = df[df['faturamento'] >2000] #importante para prova!
df_filter
```

Região Sudeste + produto Camiseta


```python
df_filter = df[ (df['regiao']=='Sudeste') & (df['produto']=='Camiseta')]
df_filter
```

Maior Faturamento

```python
df['faturamento'].max()
```

Top 5 Maiores Vendas (Faturamento)

```python
df_top5 = df.sort_values('faturamento',ascending=True).head(5)
df_top5
```

Media De faturamento por Mês

```python
df_filter = df.groupby('mes')['faturamento'].mean().reset_index()
df_filter
```

Soma De faturamento Por mês

```python
df_filter = df.groupby('mes')['faturamento'].sum().reset_index()
df_filter
```

Faturamento Total Por Produto

```python
df_filter = df.groupby('produto')['faturamento'].sum().reset_index()
df_filter
```

Faturamento Total por região (ordenado)

```python
df_filter = df.groupby ('regiao')['faturamento'].sum().reset_index().sort_values('faturamento',ascending = False).head(6)
df_filter
```

Produto com maior quantidade total vendido 

```python
df_filter = df.groupby('produto')['qtd'].sum(0).reset_index().head().sort_values('qtd',ascending=False)
df_filter
```
